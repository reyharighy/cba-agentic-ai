# Agent graph → Mermaid / PNG

This notebook builds `Graph` the same way the API does, then uses LangGraph’s built-in helpers:

- `compiled.get_graph()` → `langchain_core.runnables.graph.Graph`
- `.draw_mermaid()` → Mermaid syntax string
- `.draw_mermaid_png()` → PNG bytes

**Outputs** (under the repository’s `notebook/` folder):

- `agent_graph_mermaid.txt` — Mermaid syntax
- `Binary-Responsibility Agent Graph.png` — diagram from `draw_mermaid_png()`

`ROOT` and `NOTEBOOK_DIR` are inferred from the current working directory (you can run the notebook from the repo root or from the `notebook/` directory).

```python
from pathlib import Path
import sys

_here = Path.cwd().resolve()

if (_here / "agent").is_dir():
    ROOT = _here
    NOTEBOOK_DIR = ROOT / "notebook"
elif _here.name == "notebook" and (_here.parent / "agent").is_dir():
    ROOT = _here.parent
    NOTEBOOK_DIR = _here
else:
    ROOT = _here.parent
    NOTEBOOK_DIR = ROOT / "notebook"

NOTEBOOK_DIR.mkdir(parents=True, exist_ok=True)
sys.path.insert(0, str(ROOT))

from agent.graph import Graph

compiled = Graph().build_graph()
graph = compiled.get_graph()

mermaid = graph.draw_mermaid()
print(mermaid)

out_txt = NOTEBOOK_DIR / "agent_graph_mermaid.txt"
out_txt.write_text(mermaid, encoding="utf-8")
print("Wrote", out_txt)
```

```python
png = graph.draw_mermaid_png()
out_png = NOTEBOOK_DIR / "Binary-Responsibility Agent Graph.png"
out_png.write_bytes(png)
print("Wrote", out_png, "—", len(png), "bytes")
```
